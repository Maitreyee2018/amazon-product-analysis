### Data Cleaning

```python
df['product_id'].nunique()
```

    product_id has duplicates. hence, dropping duplicate product_id 

```python
df=df.drop_duplicates(subset='product_id')
print(df.shape)
```

    cleaning symbols in the numeric columns - discounted_price,actual_price,discount_percentage,rating,rating_count 

```python
df['discounted_price'] = (
    df['discounted_price']
    .str.replace('₹','',regex=False)
    .str.replace(',','',regex=False)
)
```

```python
numeric_discounted_price = pd.to_numeric(df['discounted_price'], errors='coerce')
non_numeric_mask=numeric_discounted_price.isna()
non_numeric_discounted_price=df[non_numeric_mask]['discounted_price']
print(non_numeric_discounted_price)
```

```python
df['discounted_price']=pd.to_numeric(df['discounted_price'], errors='coerce')
```

```python
df['actual_price'] = (
    df['actual_price']
    .str.replace('₹','',regex=False)
    .str.replace(',','',regex=False)
)
```

```python
numeric_actual_price = pd.to_numeric(df['actual_price'], errors='coerce')
non_numeric_mask=numeric_actual_price.isna()
non_numeric_actual_price=df[non_numeric_mask]['actual_price']
print(non_numeric_actual_price)
```

```python
df['actual_price']=pd.to_numeric(df['actual_price'], errors='coerce')
```

```python
df['discount_percentage'] = (
    df['discount_percentage']
    .str.replace('%','',regex=False)
    .str.replace(',','',regex=False)
)
```

```python
numeric_discounted_per = pd.to_numeric(df['discount_percentage'], errors='coerce')
non_numeric_mask=numeric_discounted_per.isna()
non_numeric_discounted_per=df[non_numeric_mask]['discount_percentage']
print(non_numeric_discounted_per)
```

```python
df['discount_percentage']=pd.to_numeric(df['discount_percentage'], errors='coerce')
```

```python
numeric_rating = pd.to_numeric(df['rating'], errors='coerce')
non_numeric_mask=numeric_rating.isna()
non_numeric_rating=df[non_numeric_mask]['rating']
print(non_numeric_rating)
```

```python
df[non_numeric_mask]
```

    Only 1 row with non numeric rating value hence can be dropped

```python
df['rating']=pd.to_numeric(df['rating'], errors='coerce')
df=df.dropna(subset=['rating'])
print(df.shape)
```

```python
df['rating_count'] = (
    df['rating_count']
    .str.replace('%','',regex=False)
    .str.replace(',','',regex=False)
)
```

```python
numeric_rating_count = pd.to_numeric(df['rating_count'], errors='coerce')
non_numeric_mask=numeric_rating_count.isna()
non_numeric_rating_count=df[non_numeric_mask]['rating_count']
print(non_numeric_rating_count)
```

```python
df['rating_count']=pd.to_numeric(df['rating_count'], errors='coerce')
df=df.dropna(subset=['rating_count'])
```

```python
df['rating_count']=pd.to_numeric(df['rating_count'], errors='coerce')
```

### Data Analysis

#### Top 10 rated category based on (rating & rating_count)

```python
weighted_category = (
    df.groupby('category').apply(lambda x: (x['rating'] *
                                x['rating_count']).sum() /
                                x['rating_count'].sum() )
                          .sort_values(ascending=False).head(10)
)

print(weighted_category)
```

#### Top 10 discounted category based on (discount_percentage)

```python
df.groupby('category')['discount_percentage'].mean().sort_values(ascending=False).head(10)
```

```python
df.groupby('rating')['discount_percentage'].mean().sort_values(ascending=False).head(10)
```

```python
df['discount_percentage'].corr(df['rating'])
```

    Higher discounts is slightly associated with lower ratings but effect is weak

#### Discounted Price impact on Rating?

```python
plt.scatter(x=df['discounted_price'].sort_values(ascending=False), y=df['rating'], color='blue', marker='o', s=100, alpha=0.7, edgecolors='black')
plt.title("Price vs Rating")
plt.xlabel("Discounted Price")
plt.ylabel("Rating")
plt.show()
```

```python
df['discounted_price'].corr(df['rating'])
```

    Price does not influence rating
    Customer rate based on product quality expectation and not price alone
    Rating distribution is compressed which is common in e-commerce
    Price is not a strong driver for rating.    

#### Discounted Price impact on No of reviews?

```python
plt.scatter(x=df['discounted_price'].sort_values(ascending=False), y=df['rating_count'], color='blue', marker='o', s=100, alpha=0.7, edgecolors='black')
plt.title("Price vs Review")
plt.xlabel("Discounted Price")
plt.ylabel("Rating Count")
plt.show()
```

```python
df['discounted_price'].corr(df['rating_count'])
```

    No relationship between Price vs Rating volume, popularity depends on category,brand,necessity.
    Expensive products are necessarily more or less reviewed

### Feature Engineering

#### Find best value products

```python
df['rating_count'].astype(float)
```

```python
df['value_score'] = df ['rating'] * df['discount_percentage'] * np.log1p(df['rating_count'])
df.sort_values('value_score',ascending=False).head(10)
```

#### Keyword Frquency

```python
from collections import Counter
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS

words=''.join(df['review_content'].astype(str)).lower().split()
filtered=[w for w in words if w not in ENGLISH_STOP_WORDS  and len(w)>2]
Counter(filtered).most_common(20)
```

#### Complain Themes

```python
low = "".join(df[df['rating']<=3]
['review_content'].astype(str)).lower()
high = "".join(df[df['rating']>=4]
['review_content'].astype(str)).lower()
```

```python
words=low.split()
low_filtered=[w for w in words if w not in ENGLISH_STOP_WORDS  and len(w)>2]
Counter(low_filtered).most_common(30)
```

    Lower ratings contains review content text with heating, quality, replacement, etc

#### Praise Themes

```python
words=high.split()
high_filtered=[w for w in words if w not in ENGLISH_STOP_WORDS  and len(w)>2]
Counter(high_filtered).most_common(20)
```

    Higher ratings contains review content text with good, quality, use, etc

### Final Conclusion

    Product rating are clustered between 3.5 to 4.5
    Correlation between discounted price and rating is weak
    Indicates that Customer satisfaction is influenced by product quality and expectation and NOT by price
    Price alone do not drive rating.
    
    Highly discounted products do not necessarily receive better reviews
    Discounts might be used for inventory clearance or competitive pricing only (not by quality)
    
    Expensive products are not inherently more/less popular. Product category and necessity is the driver for popularity.
    Customers short/long reviews do not indicate satisfaction level 
    Low rated reviews frquently mentioned about products not prove to be as useful, refund, replacement, defective, not heating, not working
    High rated reviews frquently mentioned good product quality, liked this product, easy to use

#### Value discovery

    Combining rating,discount percentage, rating count
    Best value product - B014I8SX4Y,Amazon Basics High-Speed HDMI Cable
    Underrated product - found 8 such underrated product based on high rating low visibility
    Category level pricing strategies can be decided based on findings of top 10 rated category analysis
    

#### Business Insights

    Ratings are relatively stable across price ranges
    Discounts alone do not improve customer satisfaction
    Customer reviews are influenced by product experience and not by pricing tactics 
    Integrating pricing, engagement and review text provide more complete view of product performance
    
    This analysis demonstrates how combining pricing metrics, engagement signals, customer feedback can support data driven product strategy and recommendation systems in e-commerce platform.
